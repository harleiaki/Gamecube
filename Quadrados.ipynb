```python
import pygame
import random

# Define as dimensões da janela do jogo
WIDTH = 800
HEIGHT = 600

# Define as cores usadas no jogo
WHITE = (255, 255, 255)
BLACK = (0, 0, 0)
RED = (255, 0, 0)

# Define a classe do jogador
class Player(pygame.sprite.Sprite):
    def __init__(self):
        super().__init__()
        self.image = pygame.Surface((50, 50))
        self.image.fill(WHITE)
        self.rect = self.image.get_rect()
        self.rect.center = (WIDTH / 2, HEIGHT - 50)

    def update(self):
        # Move o jogador com as setas do teclado
        keys = pygame.key.get_pressed()
        if keys[pygame.K_LEFT]:
            self.rect.x -= 5
        if keys[pygame.K_RIGHT]:
            self.rect.x += 5

        # Mantém o jogador dentro da tela
        if self.rect.left < 0:
            self.rect.left = 0
        elif self.rect.right > WIDTH:
            self.rect.right = WIDTH

# Define a classe do inimigo
class Enemy(pygame.sprite.Sprite):
    def __init__(self):
        super().__init__()
        self.image = pygame.Surface((30, 30))
        self.image.fill(RED)
        self.rect = self.image.get_rect()
        self.rect.x = random.randrange(WIDTH - self.rect.width)
        self.rect.y = random.randrange(-100, -40)
        self.speedy = random.randrange(1, 8)

    def update(self):
        # Move o inimigo para baixo
        self.rect.y += self.speedy
        if self.rect.top > HEIGHT + 10:
            self.rect.x = random.randrange(WIDTH - self.rect.width)
            self.rect.y = random.randrange(-100, -40)
            self.speedy = random.randrange(1, 8)

# Inicia o Pygame
pygame.init()

# Cria a janela do jogo
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Meu Jogo")

# Cria os sprites
all_sprites = pygame.sprite.Group()
player = Player()
all_sprites.add(player)
enemies = pygame.sprite.Group()
for i in range(8):
    enemy = Enemy()
    all_sprites.add(enemy)
    enemies.add(enemy)

# Cria o relógio para limitar o FPS
clock = pygame.time.Clock()

# Loop principal do jogo
running = True
while running:
    # Limita o FPS
    clock.tick(60)

    # Processa eventos
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    # Atualiza os sprites
    all_sprites.update()

    # Verifica se o jogador colidiu com um inimigo
    hits = pygame.sprite.spritecollide(player, enemies, False)
    if hits:
        running = False

    # Desenha os sprites na tela
    screen.fill(BLACK)
    all_sprites.draw(screen)

    # Atualiza a tela
    pygame.display.flip()

# Encerra o Pygame
pygame.quit()
```
